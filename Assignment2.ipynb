```python
# importing numpy
import numpy as np

#creating random array with 15 elements which are in between 1 and 20
vector1 = np.random.randint(1,20,15)
print("Original vector:")
print(vector1)

#Reshaping the vector to 3,5
vector2=np.reshape(vector1,(3,5))
print("Reshaped vector:")
print(vector2)

#Display shape
print("Vector shape:")
print(vector2.shape)

#Find maximum element for each row
max_elements = np.amax(vector2, axis = 1)
max_elements = max_elements[: , None]
#Replacing the max value by 0
new_arr = np.where(vector2 == max_elements, 0, vector2)
print("New array in which max elements are replaced by 0:")
print(new_arr)
```

```python
import pandas as pd
data1=pd.read_csv("data.csv")

# Statistical description about data
data1.describe()

# Replacing null values by mean 
data1=data1.fillna(data1.mean())
print("Replaced null values by mean:")
print(data1)

# Aggregate the data of columns Pulse and Calories using: min, max, count, mean
print("Min, max, count and mean values:")
print(data1[['Pulse','Calories']].agg(['min','max','mean','count']))
  
# Filter dataframe by calories    
print("Filtered data by Calories:")
df1=data1[(data1['Calories']>500) & (data1['Calories']<1000)]
print(df1)

# Filter dataframe by calories and pulse
df2=data1[(data1['Calories']>500) & (data1['Pulse']<100)]
print("Filtered data by Calories and Pulse:")
print(df2)

# Dataframe that contains all the columns from data except column “Maxpulse”
df_modified = pd.DataFrame(data1, columns = ['Duration', 'Pulse', 'Calories'])
print("df_modified:")
print(df_modified.head())

# Delete the “Maxpulse” column 
del data1['Maxpulse']
print("After deleting column Maxpulse:")
print(data1.head())

# Convert datatype of Calories column to int
data1['Calories'] = data1['Calories'].astype("int")
print("Updated data type of Calories column to int:")
print(data1['Calories'].dtypes)

# Scatter plot for the two columns Duration and Calories
plt1=data1.plot.scatter(x='Duration',y='Calories')
```

```python
#3. Matplotlib
#1. Write a Python programming to create a below chart of the popularity of programming Languages.

#importing matplotlib
from matplotlib import pyplot as plt

# Popularity as slices
slices = [ 22.2, 17.6, 8.8, 8, 7.7, 6.7]

# Programming languages as labels
labels = [ 'Java', 'Python','PHP','JavaScript','C#','C++']

# Emphasis on a cut for Java
explode = [0.1, 0, 0, 0,0, 0]

# Plotting on a pie chart
# autopct - To calculate percentages
# wedgeprops - To set the borders
plt.pie(slices, labels=labels, explode=explode, shadow=True,autopct=lambda p:f'{p:.1f}%',
        wedgeprops={'edgecolor': 'black'},startangle=140)

# Title to pie chart
plt.title('Popularity of programming languages')

# To adjust subplot params so that the subplot fits in to the figure area
plt.tight_layout()

#Display piechart
plt.show()
```
